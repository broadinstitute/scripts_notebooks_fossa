```python
import pandas as pd
import numpy as np
import pycytominer
import easygui as eg
import os
import sys

def stringToBool(correlation_input):
    if correlation_input == 'yes':
       return True
    elif correlation_input == 'no':
       return False
    else:
      raise ValueError

%load_ext autoreload
%autoreload 2
```

```python
def add_prefix_compartments(compartments, path, day, list_cols):

    df_all = []
    for cmpt in compartments:
        df = pd.read_csv(rf"{path}\{day}\{cmpt}.csv", low_memory=False)
        # print(df)
        #pop metadata and save it to use later
        list_to_pop = ['Metadata_Cell', 'Metadata_Protein', 'Metadata_Time', 'Metadata_Treatment']
        meta = [df.pop(col) for col in list_to_pop]
        df_meta = pd.concat(meta, axis=1)
        list_df = list(df.columns)
        #pop all metadata, filename and pathname cols
        for name in list_cols:
            for cols in range(len(list_df)):
                if name in list_df[cols]:
                    df.drop(list_df[cols], axis='columns', inplace=True)
        
        list_df = list(df.columns)
        #here we just add Nuclei, Cell or Cytoplasm before each col name
        for cols in range(len(list_df)):
            df.rename(columns={list_df[cols]: cmpt+"_"+list_df[cols]}, inplace=True, errors='raise') 
        df_after = pd.concat([df_meta,df], axis = 'columns')
        # features = pycytominer.cyto_utils.infer_cp_features(df_after,compartments=compartments)
        df_all.append(df_after)
    df_join = pd.concat(df_all, axis='columns')
    df_join.insert(0,'Metadata_Plate','')
    df_join['Metadata_Plate'] = day

    return df_join
```
